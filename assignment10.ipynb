1. Write a Python program to find sum of elements in list?
2. Write a Python program to Multiply all numbers in the list?
3. Write a Python program to find smallest number in a list?
4. Write a Python program to find largest number in a list?
5. Write a Python program to find second largest number in a list?
6. Write a Python program to find N largest elements from a list?
7. Write a Python program to print even numbers in a list?
8. Write a Python program to print odd numbers in a List?
9. Write a Python program to Remove empty List from List?
10. Write a Python program to Cloning or Copying a list?
11. Write a Python program to Count occurrences of an element in a list?

```python
class ListOperations:
    def __init__(self, lst):
        self.lst = lst

    def find_sum(self):
        return sum(self.lst)

    def find_product(self):
        product = 1
        for num in self.lst:
            product *= num
        return product

    def find_smallest(self):
        if not self.lst:
            raise ValueError("List is empty.")
        return min(self.lst)

    def find_largest(self):
        if not self.lst:
            raise ValueError("List is empty.")
        return max(self.lst)

    def find_second_largest(self):
        if len(self.lst) < 2:
            raise ValueError("List should have at least two elements.")
        sorted_lst = sorted(self.lst)
        return sorted_lst[-2]

    def find_n_largest(self, n):
        if n > len(self.lst):
            raise ValueError("N is greater than the number of elements in the list.")
        sorted_lst = sorted(self.lst, reverse=True)
        return sorted_lst[:n]

    def find_even_numbers(self):
        return [num for num in self.lst if num % 2 == 0]

    def find_odd_numbers(self):
        return [num for num in self.lst if num % 2 != 0]

    def remove_empty_lists(self):
        return [sub_list for sub_list in self.lst if sub_list]

    def clone_list(self):
        return self.lst.copy()

    def count_occurrences(self, element):
        return self.lst.count(element)


if __name__ == "__main__":
    try:
        # Input list from the user
        lst = list(map(int, input("Enter elements of the list (comma-separated): ").split(',')))

        list_ops = ListOperations(lst)

        # Task 1: Find sum of elements in list
        print("Sum of elements in the list:", list_ops.find_sum())

        # Task 2: Multiply all numbers in the list
        print("Product of elements in the list:", list_ops.find_product())

        # Task 3: Find smallest number in a list
        print("Smallest number in the list:", list_ops.find_smallest())

        # Task 4: Find largest number in a list
        print("Largest number in the list:", list_ops.find_largest())

        # Task 5: Find second largest number in a list
        print("Second largest number in the list:", list_ops.find_second_largest())

        # Task 6: Find N largest elements from a list
        n = int(input("Enter the value of N: "))
        print(f"{n} largest elements in the list:", list_ops.find_n_largest(n))

        # Task 7: Print even numbers in a list
        print("Even numbers in the list:", list_ops.find_even_numbers())

        # Task 8: Print odd numbers in a list
        print("Odd numbers in the list:", list_ops.find_odd_numbers())

        # Task 9: Remove empty lists from the list
        print("List after removing empty lists:", list_ops.remove_empty_lists())

        # Task 10: Cloning or Copying a list
        cloned_list = list_ops.clone_list()
        print("Cloned list:", cloned_list)

        # Task 11: Count occurrences of an element in a list
        element = int(input("Enter the element to count occurrences: "))
        print(f"Occurrences of {element} in the list:", list_ops.count_occurrences(element))

    except ValueError as e:
        print("Error:", e)
```
