11. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?

```python
# ans 11 - 
print("output for range(10)")
for i in range(10):
  print(i, end = " ")
print()
print("OUTPUT FOR RANGE(0,10)")
for j in range(0,10):
  print(j, end = " ")
print()
print("for (0,10,1)")
for k in range(0,10,1):
  print(k, end =" ")
print()
print("lets see the difference when we use 2 instead of 1")
for l in range (0, 10, 2):
  print(l, end = " " )
# range function can take three parameter range(a,b,c) , a = denotes start from here, b = denotes end here, and c= denotes step-size that help you skip value between a and b depending on the step size we provide. 
```
